```python
from pytrends.request import TrendReq

#create pytrend object to request data from google trends
pytrends = TrendReq(hl='en-us')

#make a list of keywords you wanna search in trends
keywords = ['Python', 'R', 'C++', 'Java','HTML']            
pytrends.build_payload(keywords, timeframe='today 5-y')     #5-y says last 5 years

data = pytrends.interest_over_time()                        #search vs time
data

```

```python
type(data)
```

```python
import matplotlib.pyplot as plt
plt.plot(data)

plt.title("Google Searches")
plt.xlabel("Year")
plt.ylabel("Weekly Searches")

#add legend to show which language the color represents
plt.legend(keywords)
plt.show()
```

```python
#analyze data by comparing it with python
#compare with R
focus = ['Python', 'R']
plt.plot(data[focus])
plt.legend(focus)
```

```python
#compare python with C++
focus = ['Python', 'C++']
plt.plot(data[focus])
plt.legend(focus)
```

```python
#compare python with Java
focus = ['Python', 'Java']
plt.plot(data[focus])
plt.legend(focus)
```

```python
#comapre Python with HTML
focus = ['Python', 'HTML']
plt.plot(data[focus])
plt.legend(focus)
```

```python
#search or analyze data by region
data2= pytrends.interest_by_region(resolution='COUNTRY')    #searches in region
data2=data2['Python'].nlargest(10)
data2
```

```python
#convert data to data frame
data2=data2.to_frame()
data2
```

```python
#bar chart
data2.plot(kind='bar')
plt.title("Python Searches by top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Searches")
```
